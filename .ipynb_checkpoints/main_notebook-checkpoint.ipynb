```python
import pandas as pd
from graphviz import Graph, Digraph
```

```python
df_dco = pd.read_csv("data/distrito_comuna_orga+zonas.csv", sep=",", encoding="UTF-8")
df_mcm = pd.read_csv("data/mcm_orga.csv", sep=",", encoding="UTF-8")
df_onn = pd.read_csv("data/orga_nivel_nombre.csv", sep=",", encoding="UTF-8")
df_pco = pd.read_csv("data/provincia_comuna_orga+zonas.csv", sep=",", encoding="UTF-8")
df_uco = pd.read_csv("data/universidad_campus_orga.csv", sep=",", encoding="UTF-8")
df_uo = pd.read_csv("data/universidad_orga.csv", sep=",", encoding="UTF-8")
```

```python
def GenerarGrafico3columnas(df,nombre,archivo,i,j):
    dot = Digraph(comment=nombre, filename='{}.gv'.format(archivo), engine='dot',
                  graph_attr={'rankdir':'LR','K':'4','nodesep':'0','overlap':'false','repulsiveforce':i, 'pad':'1'},
                  node_attr={'color':'lavender','fontsize':'36', 'style': 'filled'})
    for row in df.values:
        #print(row)
        if row[0] == j:
            dot.edge(str(row[0]),row[1])
            dot.edge(row[1],row[2])
    #dot.unflatten(stagger=3)
    #print(dot.source)
    dot = dot.unflatten(stagger=2)
    dot.format = 'png'
    dot.view()
    
def GenerarGrafico3ColumnasSinCondicion(df,nombre,archivo,i):
    dot = Digraph(comment=nombre, filename='{}.gv'.format(archivo), engine='dot',
                  graph_attr={'rankdir':'LR','K':'4','nodesep':'0','overlap':'false','repulsiveforce':i, 'pad':'1'},
                  node_attr={'color':'lavender','fontsize':'36', 'style': 'filled'})
    for row in df.values:
        #print(row)
        dot.edge(str(row[0]),row[1])
        dot.edge(row[1],row[2])
    #dot.unflatten(stagger=3)
    #print(dot.source)
    dot = dot.unflatten(stagger=2)
    dot.format = 'png'
    dot.view()
        
def GenerarGrafico2columnas(df,nombre,archivo):
    dot = Digraph(comment=nombre, filename='{}.gv'.format(archivo), engine='dot',
                  graph_attr={'rankdir':'LR','K':'4','nodesep':'0','overlap':'false','repulsiveforce':'1', 'pad':'1'},
                  node_attr={'color':'lavender','fontsize':'36', 'style': 'filled'})
    for row in df.values:
        print(row)
        dot.edge(str(row[0]),row[1])
    dot.format = 'png'
    dot.view()
```

```python
#GenerarGrafico3columnas(df_dco,'OrganizacionesDistritoComunaZonas','dco_dot14','1','8')

for j in range(8,15):
    GenerarGrafico3columnas(df_dco,'OrganizacionesDistritoComunaZonas','graphs/dco/dco_distrito_{}'.format(j),'1',j)
```
